```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# Create folder
os.makedirs('smallSprites', exist_ok=True)

# Base URL
base_url = 'https://archives.bulbagarden.net'
category_url = f'{base_url}/wiki/Category:Sword_and_Shield_menu_sprites'

while category_url:
    print(f'Processing: {category_url}')
    response = requests.get(category_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    thumbnails = soup.select('.gallerybox .thumb img')

    for img in thumbnails:
        thumb_url = img.get('src')
        if not thumb_url:
            continue

        full_img_url = 'https:' + thumb_url if thumb_url.startswith('//') else thumb_url
        filename = os.path.basename(full_img_url)

        img_data = requests.get(full_img_url).content
        with open(os.path.join('smallSprites', filename), 'wb') as f:
            f.write(img_data)
        print(f'Downloading: {filename}')

    next_link = soup.find('a', text='next page')
    if next_link:
        category_url = urljoin(base_url, next_link['href'])
    else:
        category_url = None

print('Download complete.')
```
